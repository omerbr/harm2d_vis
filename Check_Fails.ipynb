```python
def diff_sasha(f,dir,kind="central"):
    #convert direction into axis
    if len(f.shape)==2: #2D matrix dir=0,1
        axis = dir
    elif len(f.shape)>=3: #3D or 4D matrix, for dir=2, axis=-2 = 2(4D) or 1 (3D)
        axis = dir-4
    else: #1D 
        axis = 0
    #number of cells in the dir direction
    n = f.shape[axis]
    #if the dimension does not exist, return zero
    if 1 == n:
        return(0*f)
    #move dir to be the zeroth direction
    f1 = np.moveaxis(f,axis,0)
    diff1 = np.zeros_like(f1)
    if kind == "central":
        #take central difference in the zeroth direction 
        diff1[1:-1] = 0.5*(f1[2:]-f1[:-2])
        #fill the edge cells with extrapolated derivative
        diff1[0] = (f1[1]-f1[0])/2 #diff1[1]
        diff1[-1] = (f1[-1]-f1[-2])/2 #diff1[-2]
    elif kind == "right" or kind == "forward":
        #take right difference in the zeroth direction
        diff1[0:-1] = (f1[1:]-f1[:-1])
        #fill the edge cells with extrapolated derivative
        diff1[-1] = (2-f1[-1])#diff1[-2]/2
    elif kind == "left" or kind == "backward":
        #take right difference in the zeroth direction
        diff1[1:] = (f1[1:]-f1[:-1])
        #fill the edge cells with extrapolated derivative
        diff1[0] = diff1[1]/2
    else:
        print("Unknow derivative type: %s" % kind)
    diff = np.moveaxis(diff1,0,axis)
    return(diff)

#note: curl is taken assuming that spatial directions are 1,2,3 both in vec and output
def curl_sasha(v, gdet = 1., dx = [1.,1.,1.,1.], kind = "central"):
    curl = np.zeros_like(v)
    if kind == "central":
        curl[1] = (diff_sasha(v[3],2)/dx[2] - diff_sasha(v[2],3)/dx[3])/gdet
        curl[2] = (diff_sasha(v[1],3)/dx[3] - diff_sasha(v[3],1)/dx[1])/gdet
        curl[3] = (diff_sasha(v[2],1)/dx[1] - diff_sasha(v[1],2)/dx[2])/gdet
    elif kind == "fluxct":
        k = "forward"
        curl[1] = (diff_sasha(avg(v[3],1),2,k)/dx[2] - diff_sasha(avg(v[2],1),3,k)/dx[3])/gdet
        curl[2] = (diff_sasha(avg(v[1],2),3,k)/dx[3] - diff_sasha(avg(v[3],2),1,k)/dx[1])/gdet
        curl[3] = (diff_sasha(avg(v[2],3),1,k)/dx[1] - diff_sasha(avg(v[1],3),2,k)/dx[2])/gdet
    return(curl)
```
